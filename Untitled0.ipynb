```python
import pandas as pd
from datetime import datetime

```

```python
data = pd.read_csv('/train.csv')
```

```python
data.head()
```

```python
missing_values = data.isnull().sum()
```

```python
print("Missing values in each column:")
```

```python
print(missing_values)
```

```python
data.head()
```

```python
data['Mileage'] = data['Mileage'].str.replace(r'\D', '', regex=True)
```

```python
data['Mileage'] = pd.to_numeric(data['Mileage'], errors='coerce')
```

```python
data['Mileage'].fillna(data['Mileage'].mean(), inplace=True)
```

```python
data['Engine'] = data['Engine'].str.replace(r'\D', '', regex=True)
```

```python
data['Engine'] = pd.to_numeric(data['Engine'], errors='coerce')
```

```python
data['Engine'].fillna(data['Engine'].mean(), inplace=True)
```

```python
data['Power'] = data['Power'].str.replace(r'\D', '', regex=True)
```

```python
data['Power'] = pd.to_numeric(data['Power'], errors='coerce')
```

```python
data['Power'].fillna(data['Power'].mean(), inplace=True)
```

```python
data['Seats'].fillna(data['Seats'].mean(), inplace=True)
```

```python
data.drop(columns=['New_Price'], inplace=True)
```

```python
updated_missing_values = data.isnull().sum()
```

```python
print("\nUpdated missing values in each column:")
```

```python
print(updated_missing_values)
```
