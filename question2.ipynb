```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'StudentsPerformance.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Visualization 1: Histogram of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], bins=20, kde=True, color='skyblue')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Histogram of Math Scores')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score', data=df, hue='gender', palette='Set1', alpha=0.7)
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.title('Scatter Plot of Reading vs. Writing Scores')
plt.grid(True)
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='parental level of education', data=df, palette='viridis',
              order=df['parental level of education'].value_counts().index)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Parental Education Level')
plt.ylabel('Count')
plt.title('Bar Chart of Parental Education Levels')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='math score', data=df, palette='pastel')
plt.xlabel('Gender')
plt.ylabel('Math Score')
plt.title('Box Plot of Math Scores by Gender')
plt.grid(axis='y')
plt.show()
```

```python
lunch_counts = df['lunch'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(lunch_counts, labels=lunch_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Pie Chart of Lunch Types')
plt.show()
```
